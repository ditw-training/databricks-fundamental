# Welcome

**Databricks Fundamentals — 1-Day Training**

Welcome to the training! This opening module covers the agenda, learning objectives, and environment setup. By the end of the day you will be able to navigate the Databricks workspace, load and explore data, perform basic transformations, save results to Delta Lake, and run a simple scheduled Workflow.

**Target Audience:** Data analysts, data engineers, BI specialists, and technical staff starting their journey with Databricks.

**Prerequisites:** Basic SQL knowledge, general understanding of databases and data analysis.


## Learning Objectives

After completing this training you will be able to:

- **Navigate** the Databricks workspace — notebooks, clusters, DBFS, and Catalog Explorer
- **Explain** the Lakehouse architecture and the role of Delta Lake as the default storage format
- **Understand** Unity Catalog — catalogs, schemas, and basic permissions
- **Load** data from CSV and JSON files using `spark.read`
- **Explore** data using `display()`, `show()`, `describe()`, and `summary()`
- **Apply** basic PySpark transformations: `withColumn`, `when`, `cast`
- **Manage costs** — understand cluster types, autoscaling, and efficient resource usage
- **Use** the Databricks GUI: markdown, mixing SQL and Python, simple visualizations, AI Assistant
- **Create** a simple Workflow in Databricks — parameterization and scheduling
- **Build** a basic end-to-end pipeline: load data → transform → save to Delta Lake → schedule with Workflows
