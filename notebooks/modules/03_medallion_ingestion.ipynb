## CSV Import (Customers)

> **Definition (Databricks):** `spark.read.format("csv")` — a `DataFrameReader` method that reads delimited text files. `header=true` treats the first row as column names; `inferSchema=true` causes Spark to read the file twice — once to detect types, once to load data (expensive on large datasets). The recommended production approach is to provide an explicit schema (`StructType`), which eliminates the extra scan and guarantees data types.

Loading CSV files into DataFrames using `spark.read` with schema inference and manual schema definition. We'll work with customer data throughout this section.

---

### Auto Schema Inference

Using `inferSchema=true` to let Spark detect column types automatically.


````python
# Load CSV data with automatic schema inference
customers_auto_df = (
    spark.read
    .format("csv")
    .option("header", "true")       # First line contains column names
    .option("inferSchema", "true")  # Automatic data type inference
    .load(CUSTOMERS_CSV)
)
````

````python
# Verify loaded data
customers_auto_df.printSchema()
display(customers_auto_df.limit(5))
````

### Manual Schema Definition

Defining an explicit `StructType` schema for type safety and performance.

````python
# Schema definition for customers
# Structure: customer_id (string), first_name (string), last_name (string), email (string), city (string), country (string), registration_date (timestamp)
customers_schema = StructType([
    StructField("customer_id", StringType(), nullable=False),
    StructField("first_name", StringType(), nullable=True),
    StructField("last_name", StringType(), nullable=True),
    StructField("city", StringType(), nullable=True),
    StructField("email", StringType(), nullable=True),
    StructField("country", StringType(), nullable=True),
    StructField("registration_date", TimestampType(), nullable=True)
])
````

````python
# Load CSV data with manually defined schema
customers_df = (
    spark.read
    .format("csv")
    .option("header", "true")
    .schema(customers_schema)
    .load(CUSTOMERS_CSV)
)
````

````python
# Verify loaded data
customers_df.printSchema()
display(customers_df.limit(5))
````

## Performance: inferSchema vs Manual Schema

> **Definition (Databricks):** `inferSchema=true` causes Spark to perform **two full scans** of the data source: the first to infer column types, the second to load data. With an explicit `StructType`, Spark performs **one scan** according to the provided schema. Databricks recommends explicit schemas in production environments — especially when reading from cloud object stores (S3, ADLS), where every scan generates additional I/O operation costs and execution time.

Comparing read performance between automatic schema inference and explicit schema definition to demonstrate why manual schemas are recommended in production.

---


````python
# Automatic schema inference
start_auto = time.time()
df_auto = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load(CUSTOMERS_CSV)
)
count_auto = df_auto.count()  # Action - forces execution
time_auto = time.time() - start_auto
````

````python
# Manual schema definition
start_manual = time.time()
df_manual = (
    spark.read
    .format("csv")
    .option("header", "true")
    .schema(customers_schema)
    .load(CUSTOMERS_CSV)
)
count_manual = df_manual.count()  # Action - forces execution
time_manual = time.time() - start_manual
````

````python
# Comparison
speedup = (time_auto - time_manual) / time_auto * 100
print(f"\nConclusion: Manual schema is faster by {speedup:.1f}%")
````

## DataFrame Transformations

Core PySpark transformations for data engineering: selecting columns, adding/renaming columns, type casting, deduplication, and sorting.

---

### select — Choosing Columns

> **Definition (Databricks):** `DataFrame.select(*cols)` — column projection returning a new DataFrame containing only the specified columns or expressions. Equivalent to the `SELECT` clause in SQL. Arguments can be column names (string), `Column` objects (`col()`), expressions and functions. **Does not modify** the original DataFrame — every PySpark transformation is immutable.

Projecting specific columns and applying expressions in the `select()` method.


````python
# Select specific columns
customers_selected = customers_df.select("customer_id", "first_name", "last_name", "email")
display(customers_selected.limit(5))

# Select with column expressions
customers_transformed = customers_df.select(
    col("customer_id"),
    upper(col("first_name")).alias("first_name_upper"),
    col("email")
)
display(customers_transformed.limit(5))
````

### withColumn — Adding/Modifying Columns

> **Definition (Databricks):** `DataFrame.withColumn(colName, col)` — returns a new DataFrame by adding a column or replacing an existing column with the same name. `colName` is the new column name and `col` is a `Column` expression. The method is **lazy** — it does not perform computations until an action (`count()`, `show()`, `write`) is called. Chaining multiple `withColumn()` calls is valid, though for many operations using `select()` with multiple expressions at once is more efficient.

Creating new columns or modifying existing ones using expressions.


````python
# Add a new column
customers_with_fullname = customers_df.withColumn(
    "full_name", 
    concat(col("first_name"), lit(" "), col("last_name"))
)
display(customers_with_fullname.select("customer_id", "first_name", "last_name", "full_name").limit(5))

# Add multiple columns
customers_enriched = customers_auto_df \
    .withColumn("email_lower", lower(col("email"))) \
    .withColumn("registration_year", year(col("registration_date")))
    
display(customers_enriched.select("customer_id", "email", "email_lower", "registration_date", "registration_year").limit(5))
````

### cast — Type Conversion

> **Definition (Databricks):** `Column.cast(dataType)` — converts the column data type to the specified type (e.g. `StringType`, `IntegerType`, `DateType`). If conversion is not possible (e.g. `"abc"` → `IntegerType`), the result is `null` (default `PERMISSIVE` mode). The type can be given as a `DataType` object from `pyspark.sql.types` or as a SQL string (`"int"`, `"date"`, `"timestamp"`). A null-safe alternative is `try_cast()` (available in Spark SQL).

Converting column data types using `cast()`.


````python
# Cast customer_id to string
customers_casted = customers_df.withColumn("customer_id_str", col("customer_id").cast(StringType()))
customers_df.select("customer_id").printSchema()
customers_casted.select("customer_id_str").printSchema()

# Cast timestamp to date
customers_date = customers_df.withColumn("registration_date_only", col("registration_date").cast(DateType()))
display(customers_date.select("customer_id", "registration_date", "registration_date_only").limit(5))
````

### withColumnRenamed — Renaming

> **Definition (Databricks):** `DataFrame.withColumnRenamed(existing, new)` — returns a new DataFrame with the `existing` column renamed to `new`. Does not change data content or types. When renaming multiple columns, `select()` with `col("old").alias("new")` is more efficient because it generates a single query-plan step instead of chained `withColumnRenamed` calls.

Renaming columns using `withColumnRenamed()` or `select()` with `alias()`.


````python
# Rename single column
customers_renamed = customers_df.withColumnRenamed("customer_id", "id")
display(customers_renamed.limit(5))

# Rename multiple columns using select with alias
customers_multi_renamed = customers_df.select(
    col("customer_id").alias("id"),
    col("first_name").alias("fname"),
    col("last_name").alias("lname"),
    col("email"),
    col("city")
)
display(customers_multi_renamed.limit(5))
````

### drop — Removing Columns

> **Definition (Databricks):** `DataFrame.drop(*cols)` — returns a new DataFrame without the specified columns. Multiple names can be passed as separate arguments. If a specified column does not exist in the DataFrame, the operation is silently ignored. `drop()` is complementary to `select()` — useful when you want to remove a few columns from a wide DataFrame instead of listing all columns to keep.

Removing unwanted columns from a DataFrame.


````python
# Drop single column
customers_dropped = customers_df.drop("email")

# Drop multiple columns
customers_minimal = customers_df.drop("email", "city", "country")
display(customers_minimal.limit(5))
````

### distinct / dropDuplicates — Unique Rows

> ** Definicja (Databricks):**
> - `DataFrame.distinct()` — returns a new DataFrame containing only unique rows (equivalent to `SELECT DISTINCT *` in SQL). Compares **all** columns.
> - `DataFrame.dropDuplicates(subset)` — removes duplicate rows based on the specified column subset (`subset`). If `subset` is not provided, behaves like `distinct()`. On large datasets both operations require a shuffle (repartitioning across the network) — they can be expensive.

Removing duplicate rows entirely or based on specific columns.


````python
# Get distinct countries
distinct_countries = customers_df.select("country").distinct()
display(distinct_countries.orderBy("country"))

# Drop duplicates based on specific columns
unique_locations = customers_df.select("city", "country").dropDuplicates(["city", "country"])
display(unique_locations.orderBy("country", "city").limit(10))
````

### orderBy — Sorting

> **Definition (Databricks):** `DataFrame.orderBy(*cols, **kwargs)` (alias: `sort()`) — returns a new DataFrame sorted by the specified columns. Default direction is ascending (`asc`); descending is achieved with `desc()` or `col("name").desc()`. **Production note:** `orderBy()` requires a global sort (full shuffle of all partitions to a single node), which is expensive on large datasets. In pipelines it is recommended to defer sorting to the last step or use `sortWithinPartitions()` when local sorting is sufficient.

Sorting rows by one or more columns in ascending or descending order.


````python
# Sort by single column (ascending)
customers_sorted_asc = customers_df.orderBy("registration_date")
display(customers_sorted_asc.select("customer_id", "first_name", "last_name", "registration_date").limit(5))

# Sort by multiple columns with different directions
customers_sorted_multi = customers_df.orderBy(asc("country"), desc("registration_date"))
display(customers_sorted_multi.select("customer_id", "first_name", "country", "registration_date").limit(10))
````

## Filtering Data

Filtering rows using conditions, multiple predicates, `isin()`, null handling, and string pattern matching.

---

### Simple Conditions

> ** Definicja (Databricks):** `DataFrame.filter(condition)` (alias: `where()`) — returns a new DataFrame containing only rows that satisfy the given logical condition. The condition can be a `Column` expression (e.g. `col("country") == "USA"`) or a SQL string (e.g. `"country = 'USA'"`). `filter()` and `where()` are **identical** — `where()` is an alias added for users accustomed to SQL syntax. Filtering uses *predicate pushdown* — Spark pushes it as early as possible into data reading.

`filter()` and `where()` are equivalent.


````python
# Filter by country
usa_customers = customers_auto_df.filter(col("country") == "USA")
display(usa_customers.select("customer_id", "first_name", "last_name", "country").limit(5))

# Filter using where (equivalent to filter)
nyc_customers = customers_auto_df.where(col("city") == "New York")
display(nyc_customers.select("customer_id", "first_name", "city", "country").limit(5))
````

### Multiple Conditions (`&` = AND, `|` = OR)

> ** Definicja (Databricks):** Complex conditions are combined with bitwise operators: `&` (AND), `|` (OR), `~` (NOT). **Each expression must be wrapped in parentheses**, because bitwise operators have higher precedence than comparison operators — without parentheses `col("a") == 1 & col("b") == 2` will be mis-interpreted. Alternatively use `pyspark.sql.functions.expr("a = 1 AND b = 2")` or SQL strings in `filter("a = 1 AND b = 2")`.

Combining conditions with logical operators — remember to wrap each condition in parentheses.


````python
# AND condition
usa_2023 = customers_df.filter(
    (col("country") == "USA") & (year(col("registration_date")) == 2023)
)
display(usa_2023.select("customer_id", "first_name", "country", "registration_date").limit(5))

# OR condition
usa_or_uk = customers_df.filter(
    (col("country") == "USA") | (col("country") == "UK")
)
display(usa_or_uk.select("customer_id", "first_name", "country").limit(5))
````

### isin — Filter by List

> ** Definicja (Databricks):** `Column.isin(*cols)` — returns a boolean expression `true` if the column value is in the provided list. Equivalent to the SQL `WHERE column IN (v1, v2, ...)` clause. You can pass a list as `col.isin(["a", "b"])` or unpacked values `col.isin("a", "b")`. Negation: `~col("country").isin(list)` or `col("country").isin(list) == False`.

Filtering rows where a column value matches any item in a list.


````python
# Filter by list of countries
selected_countries = ["USA", "UK", "Germany", "France"]
customers_selected_countries = customers_df.filter(col("country").isin(selected_countries))

# Show distribution by country
display(customers_selected_countries.groupBy("country").count().orderBy(desc("count")))
````

### Null Handling (`isNull`, `isNotNull`)

> ** Definicja (Databricks):**
> - `Column.isNull()` — returns `true` for rows where the column value is `null`. SQL equivalent: `IS NULL`.
> - `Column.isNotNull()` — returns `true` for rows with a non-null value. SQL equivalent: `IS NOT NULL`.
>
> In PySpark the comparison `col("x") == None` or `col("x") == "null"` **does not work** correctly — always use `isNull()` / `isNotNull()`. To fill null values use `DataFrame.fillna()` or `DataFrame.na.fill()`.

Filtering rows based on null or non-null values.


````python
# Filter rows where email is NOT null
customers_with_email = customers_df.filter(col("email").isNotNull())

# Filter rows where city IS null
customers_no_city = customers_df.filter(col("city").isNull())
if customers_no_city.count() > 0:
    display(customers_no_city.select("customer_id", "first_name", "city", "country").limit(5))
````

### String Operations (`like`, `contains`, `startswith`)

> ** Definicja (Databricks):**
> - `Column.like(pattern)` — matches a SQL wildcard pattern: `%` (any string), `_` (one character). SQL equivalent: `LIKE`.
> - `Column.contains(other)` — returns `true` if the column contains the substring `other`.
> - `Column.startswith(other)` / `endswith(other)` — checks prefix / suffix.
> - `Column.rlike(pattern)` — matches using a regular expression (Java regex).
>
> All string operations return `null` if the column value is `null`.

Pattern matching and string-based filtering on column values.


````python
# Filter using like (SQL-style pattern matching)
gmail_customers = customers_df.filter(col("email").like("%@gmail.com"))
display(gmail_customers.select("customer_id", "first_name", "email").limit(5))

# Filter using contains
new_cities = customers_df.filter(col("city").contains("New"))
display(new_cities.select("customer_id", "first_name", "city").limit(5))

# Filter using startswith
j_names = customers_df.filter(col("first_name").startswith("J"))
display(j_names.select("customer_id", "first_name", "last_name").limit(5))
````

## Aggregations

Computing summary statistics with `groupBy`, `agg()`, and built-in aggregate functions like `count`, `sum`, `avg`, `min`, and `max`.

---

### groupBy with count, sum, avg

> ** Definicja (Databricks):** `DataFrame.groupBy(*cols)` — groups rows by the specified columns and returns a `GroupedData` object on which you can call aggregate functions (`count()`, `sum()`, `avg()`, `min()`, `max()`) or the generic `agg()` method. SQL equivalent: `GROUP BY`. `count("*")` counts all rows (including nulls); `count("col")` counts only non-null values. Columns in `groupBy` become the group key in the resulting DataFrame.

Basic grouping and aggregation operations.


````python
# Count by country
customers_by_country = customers_df.groupBy("country").count().orderBy(desc("count"))
display(customers_by_country.limit(10))
````

````python
# Sum and average on orders
revenue_by_payment = orders_df.groupBy("payment_method").agg(
    sum("total_amount").alias("total_revenue"),
    avg("total_amount").alias("avg_order_value"),
    count("*").alias("order_count")
).orderBy(desc("total_revenue"))

display(revenue_by_payment)
````

### min / max

> **Definition (Databricks):** `min(col)` / `max(col)` — aggregate functions from `pyspark.sql.functions` returning the minimum / maximum column value in a group. They work on numeric types, dates and strings (alphabetically). When called without `groupBy` on `DataFrame.agg()`, they compute the global min/max for the entire DataFrame. `null` values are **ignored** by both functions.

Finding minimum and maximum values across the dataset.


````python
# Min and max order amounts
order_stats = orders_df.agg(
    min("total_amount").alias("min_amount"),
    max("total_amount").alias("max_amount"),
    avg("total_amount").alias("avg_amount")
)
display(order_stats)
````

### Multiple Aggregations with agg()

> ** Definicja (Databricks):** `GroupedData.agg(*exprs)` — accepts multiple aggregation expressions at once, generating **a single query-plan step** instead of separate operations. Each expression is an aggregate function call with an alias (`sum("amount").alias("total")`). You can use any function from `pyspark.sql.functions` or SQL expressions via `expr()`. This is the recommended way to compute multiple metrics in a single data pass — minimises shuffle and I/O costs.

Combining several aggregate functions in a single `agg()` call.


````python
# Example: Multiple aggregations on orders by customer
customer_stats = orders_df.groupBy("customer_id").agg(
    count("*").alias("total_orders"),
    sum("total_amount").alias("total_spent"),
    avg("total_amount").alias("avg_order_value"),
    min("total_amount").alias("min_order"),
    max("total_amount").alias("max_order")
).orderBy(desc("total_spent"))

display(customer_stats.limit(10))
````

### HAVING Equivalent (filter after groupBy)

> **Definition (Databricks):** In SQL the `HAVING` clause filters groups **after** aggregation. In PySpark the equivalent is `filter()` (or `where()`) called on a DataFrame **after** `groupBy().agg()`. Order matters: `filter()` **before** `groupBy` filters rows (equivalent to `WHERE`), while `filter()` **after** `groupBy().agg()` filters groups (equivalent to `HAVING`). Catalyst optimizer automatically optimises predicate order where possible.

Applying `filter()` after `groupBy().agg()` to replicate SQL's `HAVING` clause.


````python
# Customers with more than 5 orders
high_frequency = orders_df.groupBy("customer_id").agg(
    count("*").alias("order_count"),
    sum("total_amount").alias("total_spent")
).filter(col("order_count") > 5).orderBy(desc("order_count"))

display(high_frequency.limit(10))
````

## Temporary Views & SQL

Registering DataFrames as temporary views and querying them with SQL. Covers temp views, global temp views, and the DataFrame API vs SQL equivalence.

---

### Creating Temporary Views

> ** Definicja (Databricks):** `DataFrame.createOrReplaceTempView(name)` — registers a DataFrame as a temporary view in the current Spark session (`SparkSession`). The view stores the **query definition** (not a copy of the data) and is accessible via `spark.sql("SELECT * FROM name")` or `%sql SELECT * FROM name`. Lifetime: until the Spark session ends (cluster restart or notebook detach). **Not visible** in other notebooks or sessions. The `createOrReplace` method replaces an existing view without error.

Registering DataFrames as views for SQL access within the current SparkSession.


| View Type | Scope | Persistence | Syntax |
|---|---|---|---|

| **Global Temp View** | All sessions on cluster | Until cluster restart | `CREATE GLOBAL TEMP VIEW` → `global_temp.name` |
| **Permanent View** | Unity Catalog | Persistent | `CREATE VIEW catalog.schema.name` |

> **Exam Tip:** Temp views are NOT visible across notebooks. Views store the query definition, not data.

````python
# Create temporary views from our DataFrames

customers_df.createOrReplaceTempView("customers")
orders_df.createOrReplaceTempView("orders")
products_df.createOrReplaceTempView("products")

print("These views are available for SQL queries in this session.")
````

Executing SQL statements against registered temporary views.

````python
# Simple SELECT query
result = spark.sql("""
    SELECT 
        customer_id,
        COUNT(*) as order_count,
        SUM(total_amount) as total_spent
    FROM orders
    GROUP BY customer_id
    ORDER BY order_count DESC
    LIMIT 10
""")

display(result)
````

### Global Temporary Views

> ** Definicja (Databricks):** `DataFrame.createOrReplaceGlobalTempView(name)` — registers a DataFrame as a global temporary view, accessible to **all sessions** (`SparkSession`) on the same cluster. The global view is accessible via the qualified `global_temp.name` schema (e.g. `SELECT * FROM global_temp.my_view`). Lifetime: until cluster restart or explicit removal via `spark.catalog.dropGlobalTempView("name")`. Less common than a regular temp view — used when different notebooks/jobs on the same cluster need to share a temporary query definition.

Creating views accessible across all SparkSessions on the cluster via `global_temp` schema.

### Complex SQL (JOINs + Aggregations)

Combining JOINs with GROUP BY and HAVING in a single SQL query.

````python
# Example: Complex query with JOIN and aggregation

complex_query = spark.sql("""
    SELECT 
        c.customer_id,
        c.first_name,
        c.last_name,
        c.country,
        COUNT(o.order_id) as total_orders,
        SUM(o.total_amount) as total_spent,
        AVG(o.total_amount) as avg_order_value,
        MAX(o.total_amount) as largest_order
    FROM customers c
    LEFT JOIN orders o ON c.customer_id = try_cast(o.customer_id AS STRING)
    GROUP BY c.customer_id, c.first_name, c.last_name, c.country
    HAVING COUNT(o.order_id) > 0
    ORDER BY total_spent DESC
    LIMIT 10
""")

display(complex_query)
````

## JSON Operations

Working with complex JSON data: flattening arrays with `explode`, accessing nested fields with dot notation, and parsing JSON strings with `get_json_object` and `from_json`.

---

### explode — Flattening Arrays

> ** Definicja (Databricks):**
> - `explode(col)` — creates a new row for each element of an array (`ArrayType`) or map (`MapType`). Rows with an empty array (`[]`) or `null` are **skipped**.
> - `explode_outer(col)` — like `explode`, but rows with `null` or an empty array are **kept** with a `null` value in the result column.
>
> Both functions are from `pyspark.sql.functions`. Typical use: normalising JSON data with nested lists into tabular form. SQL equivalent: `LATERAL VIEW EXPLODE(col)`.

Converting array elements into individual rows using `explode()` and `explode_outer()`.


````python
# Create sample data with arrays to demonstrate explode
sample_data = [
    (1, "Customer A", ["product_1", "product_2", "product_3"]),
    (2, "Customer B", ["product_1"]),
    (3, "Customer C", []),  # Empty array
    (4, "Customer D", None)  # Null array
]

sample_schema = StructType([
    StructField("customer_id", IntegerType(), False),
    StructField("customer_name", StringType(), False),
    StructField("purchased_products", ArrayType(StringType()), True)
])

sample_df = spark.createDataFrame(sample_data, schema=sample_schema)
display(sample_df)
````

````python
# explode() - skips null and empty arrays
exploded_df = sample_df.select(
    "customer_id",
    "customer_name",
    explode("purchased_products").alias("product")
)
display(exploded_df)

# explode_outer() - keeps null and empty arrays
exploded_outer_df = sample_df.select(
    "customer_id",
    "customer_name",
    explode_outer("purchased_products").alias("product")
)
display(exploded_outer_df)
````

### Nested JSON — dot notation

> ** Definicja (Databricks):** Nested columns (`StructType`) are accessible via dot notation: `col("parent.child")` or `df["parent"]["child"]`. Spark automatically recognises nested structure when reading JSON. `printSchema()` shows the type hierarchy. To promote nested fields to the top level use `select("id", "address.city", "address.country")` — this operation is called *flattening*. The `struct()` function creates a new nested column from the specified columns.

Accessing fields within nested structs using `col("parent.child")` syntax.


````python
# Create sample data with nested structures
nested_data = customers_df.select(
    col("customer_id"),
    struct(
        col("first_name"),
        col("last_name"),
        col("email")
    ).alias("personal_info"),
    struct(
        col("city"),
        col("country")
    ).alias("location")
)

nested_data.printSchema()
display(nested_data.limit(3))

# Access nested fields
flattened = nested_data.select(
    "customer_id",
    col("personal_info.first_name").alias("first_name"),
    col("personal_info.email").alias("email"),
    col("location.country").alias("country")
)
display(flattened.limit(5))
````

### Parsing JSON Strings (`get_json_object`, `from_json`)

> ** Definicja (Databricks):**
> - `get_json_object(col, path)` — extracts a value from a JSON string based on a JSONPath (e.g. `"$.address.city"`). Returns `StringType`. Efficient for reading single fields.
> - `from_json(col, schema)` — parses a JSON string into a `StructType` structure according to the provided schema. Returns a nested column whose fields are accessible via dot notation. Recommended when multiple fields are needed at once.
> - `to_json(col)` — the reverse conversion: turns a `StructType` or `MapType` back into a JSON string.

Extracting values from JSON-encoded string columns.


````python
# Create sample data with JSON strings
json_string_data = [
    (1, '{"name": "John", "age": 30, "city": "New York"}'),
    (2, '{"name": "Jane", "age": 25, "city": "London"}'),
    (3, '{"name": "Bob", "age": 35, "city": "Paris"}')
]

json_df = spark.createDataFrame(json_string_data, ["id", "json_data"])
display(json_df)

# Extract fields using get_json_object
parsed_df = json_df.select(
    "id",
    get_json_object("json_data", "$.name").alias("name"),
    get_json_object("json_data", "$.age").alias("age"),
    get_json_object("json_data", "$.city").alias("city")
)
display(parsed_df)
````

## Joins

Combining DataFrames using different join types: inner, left, right, and full outer. All join types use the same Catalyst optimizer under the hood.

---

### Inner Join

> ** Definicja (Databricks):** `DataFrame.join(other, on, how="inner")` — joins two DataFrames returning **only rows with a match on both sides** of the join key. Rows without a match are discarded. Databricks Spark automatically chooses a strategy: *Broadcast Hash Join* (when one DataFrame is small, ≤ 10 MB by default), *Sort-Merge Join* (large datasets), or *Shuffle Hash Join*. The join result contains columns from both DataFrames — when names clash use qualifiers (`df1.col`) or aliases.

Returning only rows with matching keys in both DataFrames.


````python
# Inner join - customers with their orders

customers_with_orders = customers_df.join(
    orders_df,
    customers_df.customer_id == orders_df.customer_id,
    "inner"
).select(
    customers_df.customer_id,
    customers_df.first_name,
    customers_df.last_name,
    orders_df.order_id,
    orders_df.total_amount,
    orders_df.payment_method
)

display(customers_with_orders.limit(10))
````

### Left Join

> ** Definicja (Databricks):** `how="left"` (alias: `"left_outer"`) — keeps **all rows from the left** DataFrame. For rows without a match in the right DataFrame, right-side columns take `null`. Typical use: enriching a fact table with dimensions (e.g.orders + customer data), where a missing customer should not eliminate the order from results.

Keeping all rows from the left DataFrame, with nulls where no match exists on the right.


````python
# Left join - all orders with customer details (if available)

orders_with_customers = orders_df.join(
    customers_df,
    orders_df['customer_id'] == customers_df['customer_id'],
    "left"
).select(
    orders_df.order_id,
    orders_df.customer_id,
    customers_df.first_name,
    customers_df.last_name,
    orders_df.total_amount
)

display(orders_with_customers.limit(10))
````

### Right Join

> ** Definicja (Databricks):** `how="right"` (alias: `"right_outer"`) — keeps **all rows from the right** DataFrame. For rows without a match in the left DataFrame, left-side columns take `null`. It is the symmetric equivalent of Left Join — in practice rarely used, because the same effect can be achieved by swapping the DataFrame order and using Left Join.

Keeping all rows from the right DataFrame, with nulls where no match exists on the left.


````python
# Right join - all customers with their orders (if any)

customers_with_orders = orders_df.join(
    customers_df,
    orders_df['customer_id'] == customers_df['customer_id'],
    "right"
).select(
    customers_df.customer_id,
    customers_df.first_name,
    customers_df.last_name,
    orders_df.order_id,
    orders_df.total_amount
)

display(customers_with_orders.limit(10))
````

### Full Outer Join

> ** Definicja (Databricks):** `how="outer"` (alias: `"full"`, `"full_outer"`) — keeps **all rows from both** DataFrames. Where there is no match, the other side's columns take `null`. Typical use: auditing differences between systems — comparing two datasets and identifying records present in only one of them. It is the most expensive join type — requires a full shuffle of both sides.

Keeping all rows from both DataFrames, with nulls on either side where no match exists.


````python
# Full outer join - all customers and all orders

from pyspark.sql.functions import col

full_join = customers_df.join(
    orders_df,
    customers_df.customer_id == orders_df.customer_id,
    "outer"
).select(
    customers_df.customer_id.alias("cust_id"),
    orders_df.customer_id.alias("order_cust_id"),
    customers_df.first_name,
    col("order_id").alias("order_id"),
    orders_df.total_amount
)

print("\nSample with potential nulls:")
display(full_join.limit(10))
````

## read_files() — Unity Catalog Native Reader

> **Definition (Databricks):** `read_files(path, format, ...)` — a native SQL function in Databricks Runtime 13.3+ that enables reading files directly in SQL queries without creating an external table. Works with Unity Catalog **Volumes** (paths `/Volumes/catalog/schema/volume/file`). Automatically attaches file metadata (`_metadata.file_path`, `_metadata.file_name`, `_metadata.file_size`). Supported formats: `csv`, `json`, `parquet`, `avro`, `orc`, `text`, `binaryFile`. Recommended by Databricks for SQL-first workflows with UC — replaces `CREATE TABLE USING` with an external location.

| Feature | `read_files()` | `spark.read` |
|---|---|---|
| Language | SQL | Python |
| Schema inference | Automatic | Manual or `inferSchema` |
| UC integration | Native | Via path |
| Use case | SQL-first workflows | Programmatic pipelines |

```sql
-- Read CSV from a Volume
SELECT * FROM read_files('/Volumes/catalog/schema/volume/file.csv', format => 'csv', header => true);

-- Create table from files
CREATE TABLE bronze.customers AS SELECT * FROM read_files('/Volumes/.../file.csv');
```

> **Exam Tip:** `read_files()` is the recommended way to read files in SQL workflows with Unity Catalog.
