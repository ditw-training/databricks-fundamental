## Scenario

You are a data engineer at a retail company. Your manager needs a **fully automated pipeline** that turns raw JSON order files into a daily revenue report — ready for the analytics team by 06:00 every morning.

You will implement the pipeline as three independent notebooks following the **Medallion architecture**, then wire them together into a Databricks Workflow Job.

```
[Volume]  orders_batch.json
      |
      v  Task 01 — lab_task_01_bronze
      |  Read JSON + add metadata -> bronze.lab_orders
      |
      v  Task 02 — lab_task_02_silver
      |  Cast * enrich * filter -> silver.lab_orders
      |
      v  Task 03 — lab_task_03_gold
         Aggregate by date -> gold.lab_daily_orders
```

---

## Learning Objectives

By the end of this lab you will be able to:

- Implement a **3-layer Medallion pipeline** (Bronze → Silver → Gold)
- Ingest raw JSON data into a **Bronze Delta table** with metadata columns
- Clean and enrich data in a **Silver Delta table** (cast, filter, join)
- Aggregate business metrics in a **Gold Delta table** (daily revenue KPIs)
- Wire three notebooks into a **Databricks Workflow Job** with dependencies
- Schedule and run the pipeline end-to-end on a **Job Cluster**

## Layer responsibilities

| Layer | Table | Rule |
|-------|-------|------|
| **Bronze** | `bronze.lab_orders` | Raw data — only metadata columns added |
| **Silver** | `silver.lab_orders` | Typed, enriched, filtered |
| **Gold** | `gold.lab_daily_orders` | Aggregated — one row per day |

## Step 4 — Verify the Gold table in SQL Query Editor

After the job run completes, open **SQL Editor** (left sidebar → SQL Editor) and run the queries below.
Replace `retailhub_<your_name>` with your actual catalog name.

```sql
-- 1. Check the table exists and preview rows
SELECT *
FROM retailhub_<your_name>.gold.lab_daily_orders
ORDER BY order_date DESC
LIMIT 10;
```

```sql
-- 2. Row count — should match the number of distinct order dates in the source file
SELECT COUNT(*) AS total_days
FROM retailhub_<your_name>.gold.lab_daily_orders;
```

```sql
-- 3. Revenue sanity check — total revenue across all days
SELECT
    SUM(total_revenue)  AS grand_revenue,
    SUM(order_count)    AS grand_orders,
    MIN(order_date)     AS first_date,
    MAX(order_date)     AS last_date
FROM retailhub_<your_name>.gold.lab_daily_orders;
```

```sql
-- 4. Inspect Delta version history — confirm the last write came from the pipeline job
DESCRIBE HISTORY retailhub_<your_name>.gold.lab_daily_orders;
```

> **Tip:** You can also run `DESCRIBE TABLE EXTENDED retailhub_<your_name>.gold.lab_daily_orders` to see the table location, schema, and Unity Catalog metadata.
