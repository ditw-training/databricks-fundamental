# Test Runner — Databricks Fundamental

Runs all DEMO modules and workshop solution notebooks in sequence and reports PASS / FAIL per notebook.

| Cell | Scope |
|---|---|
| **Cell 2** | DEMO modules M01–M07 |
| **Cell 3** | Workshop solutions W1, W2, WFINAL (tasks 1–3) |

> **Prerequisites:** `00_setup.ipynb` must have been run at least once (catalog + schemas created).
> Each notebook is run with a **timeout of 600 seconds** (10 min). Adjust `TIMEOUT` if needed.

````python
# =============================================================================
# CELL 1 — DEMO modules: M01 → M07
# =============================================================================
TIMEOUT = 600  # seconds per notebook
BASE    = "../modules"

modules = [
    ("M01", f"{BASE}/01_platform_workspace"),
    ("M02", f"{BASE}/02_spark_architecture"),
    ("M03", f"{BASE}/03_medallion_ingestion"),
    ("M04", f"{BASE}/04_delta_fundamentals"),
    ("M05", f"{BASE}/05_orchestration_jobs"),
    ("M06", f"{BASE}/06_unity_catalog"),
    ("M07", f"{BASE}/07_aibi_dashboards"),
]

results = []

print("=" * 60)
print("  DEMO MODULES — test run")
print("=" * 60)

for name, path in modules:
    try:
        dbutils.notebook.run(path, TIMEOUT)
        status = "✅ PASS"
    except Exception as e:
        status = f"❌ FAIL  →  {str(e)[:120]}"
    results.append((name, path, status))
    print(f"  {name:4s}  {status}")

print()
passed = sum(1 for _, _, s in results if s.startswith("✅"))
failed = len(results) - passed
print(f"  Result: {passed}/{len(results)} passed  |  {failed} failed")

if failed:
    print("\n  Failed notebooks:")
    for name, path, status in results:
        if not status.startswith("✅"):
            print(f"    {name}: {path}")
            print(f"    {status}")
````

````python
# =============================================================================
# CELL 2 — Workshop SOLUTION notebooks: W1, W2, WFINAL tasks 1–3
# =============================================================================
TIMEOUT  = 600
BASE_W   = "../workshops"
BASE_LAB = "../workshops/final_labs/tasks"

workshops = [
    ("W1  WORKSHOP_ingestion_solution",  f"{BASE_W}/WORKSHOP_ingestion_solution"),
    ("W2  WORKSHOP_delta_solution",      f"{BASE_W}/WORKSHOP_delta_solution"),
    ("LAB01  Bronze layer solution",     f"{BASE_LAB}/lab_task_01_bronze_solution"),
    ("LAB02  Silver layer solution",     f"{BASE_LAB}/lab_task_02_silver_solution"),
    ("LAB03  Gold layer solution",       f"{BASE_LAB}/lab_task_03_gold_solution"),
]

results_w = []

print("=" * 60)
print("  WORKSHOP SOLUTIONS — test run")
print("=" * 60)

for name, path in workshops:
    try:
        dbutils.notebook.run(path, TIMEOUT)
        status = "✅ PASS"
    except Exception as e:
        status = f"❌ FAIL  →  {str(e)[:120]}"
    results_w.append((name, path, status))
    print(f"  {name:40s}  {status}")

print()
passed_w = sum(1 for _, _, s in results_w if s.startswith("✅"))
failed_w = len(results_w) - passed_w
print(f"  Result: {passed_w}/{len(results_w)} passed  |  {failed_w} failed")

if failed_w:
    print("\n  Failed notebooks:")
    for name, path, status in results_w:
        if not status.startswith("✅"):
            print(f"    {name}")
            print(f"    {status}")
````
